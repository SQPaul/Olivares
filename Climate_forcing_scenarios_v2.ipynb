### PP

```python
#Read PP

pp = rio.open(r"P:\Projects\Olivares\Climate_forcing\PP.nc")
pp_cdf = xr.open_dataset(r"P:\Projects\Olivares\Climate_forcing\PP.nc") 


#pp_50 = pd.read_excel(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\pp_T50.xlsx")
pp_100 = pd.read_excel(r"P:\Projects\Olivares\Scenarios\pp_sc1.xlsx")

#Read net dem
netdem_t = rio.open(r"P:\Projects\Olivares\GIS\net_dem1.tif")
netdem = netdem_t.read(1)

#Read mask
mask_tif = rio.open(r"P:\Projects\Olivares\GIS\real_mask2.tif") #windows
mask_array = mask_tif.read(1)

#Set gradient
g_pp = 0.0084 #PP gradient mm/m related to pp_avg

for i in range(len(pp_100)):
    ii = i+1
    if ii < 10000:
        value = ii/1000
        f_value = "{:.3f}".format(value)
        name = r"P:\Projects\Olivares\SPHY\input\climate_forcing\pp5\pp50000"+str(f_value) #windows       
    pp_it = pp.read(ii)
    if ii >= 6576 and ii <= 6940:
        if ii == 6727:
            pp_it = pp_it + 216
        elif ii == 6757:
            pp_it = pp_it + 208
        elif ii == 6788:
            pp_it = pp_it + 216
        elif pp_100.iloc[i,1] == 0:
            pp_it = pp_it
        else:
            pp_it = pp_it + 2.72
    maps = rio.open(
    str(name),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps.write(pp_it,1)
    maps.close()
    print(name)
```

### T2M

```python
tmax_cdf = xr.open_dataset(r"P:\Projects\Olivares\Climate_forcing\TMAX.nc")

tmax = rio.open(r"P:\Projects\Olivares\Climate_forcing\TMAX.nc")
tmin = rio.open(r"P:\Projects\Olivares\Climate_forcing\TMIN.nc")

tmax_100 = pd.read_excel(r"P:\Projects\Olivares\Scenarios\tmax_sc1.xlsx")
tmin_100 = pd.read_excel(r"P:\Projects\Olivares\Scenarios\tmin_sc1.xlsx")

#Read net dem
netdem_t = rio.open(r"P:\Projects\Olivares\GIS\net_dem1.tif")
netdem = netdem_t.read(1)

#Read mask
mask_tif = rio.open(r"P:\Projects\Olivares\GIS\real_mask2.tif") #windows
mask_array = mask_tif.read(1)

#Set gradient
g_tmp = -0.0067 #PP gradient mm/m related to pp_avg

for i in range(len(tmax_100)):
    ii = i+1
    if ii < 10000:
        value = ii/1000
        f_value = "{:.3f}".format(value)
        #t100
        #r"P:\Projects\Olivares\SPHY\input\climate_forcing\pp2\pp20000"
        
        name4 = r"P:\Projects\Olivares\SPHY\input\climate_forcing\tmax5\tmax500"+str(f_value)
        name5 = r"P:\Projects\Olivares\SPHY\input\climate_forcing\tmin5\tmin500"+str(f_value)
        name6 = r"P:\Projects\Olivares\SPHY\input\climate_forcing\tavg5\tavg500"+str(f_value)
      
    tmax_it = tmax.read(ii)    
    tmin_it = tmin.read(ii)
    tavg_it = (tmax_it+tmin_it)/2

    #Elevation adjustment
    
    if ii >= 6941 and ii <= 7305:
        if ii == 6940:
            tmax_it = tmax_it + 12.29
            tmin_it = tmin_it + 10.38
            tavg_it = (tmax_it + tmin_it)/2
        elif ii == 6971:
            tmax_it = tmax_it + 10.55
            tmin_it = tmin_it + 8.12
            tavg_it = (tmax_it+tmin_it)/2
        elif ii == 6999:
            tmax_it = tmax_it + 12.41
            tmin_it = tmin_it + 8.17
            tavg_it = (tmax_it+tmin_it)/2
        else:
            tmax_it = tmax_it+8.23
            tmin_it = tmin_it+6.56
            tavg_it = (tmax_it+tmin_it)/2
    
    #### t100
    #Tmax
    maps4 = rio.open(
    str(name4),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps4.write(tmax_it,1)
    maps4.close()
    
    #Tmin
    maps5 = rio.open(
    str(name5),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps5.write(tmin_it,1)
    maps5.close()
    
    #Tavg
    maps6 = rio.open(
    str(name6),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps6.write(tavg_it,1)
    maps6.close()

    print(name4)
    print(name5)
    print(name6)
```
